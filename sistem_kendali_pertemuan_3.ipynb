# **Create state space models**

image.png

Example 1.

image.png

```python
import numpy as np
import control as ct
```

```python
A = np.array([[0, 1], [-1, -1]])
B = np.array([[0], [1]])
C = np.array([[1, 0]])
D = np.array([[0]])

S = ct.ss(A, B, C, D)

print('State Space Model\n', S)
```

```python
#create step input
import matplotlib.pyplot as plt
%matplotlib inline

x1 = np.linspace(0, 20, 100)
y1 = 4*np.ones(100)
plt.plot(x1, y1)
plt.grid()
plt.show()
```

```python
x,y = ct.forced_response(S, x1, y1)
```

```python
plt.plot(x, y, 'green', x1, y1, 'blue')
plt.grid()
plt.show()
```

```python
H = ct.ss2tf(S)
#print('H(s)=', H)
x,y = ct.forced_response(H, x1, y1)
plt.plot(x, y, 'green', x1, y1, 'blue')
plt.grid()
plt.show()
```

```python
#solution a
A = np.array([[0, 1, 0], [0, 0, 1], [-24, -26, -9]])
B = np.array([[0], [1], [24]])
C = np.array([[1, 0, 0]])
D = np.array([[0]])

Sa = ct.ss(A, B, C, D)

print('State Space Model\n', Sa)
Ha = ct.ss2tf(Sa)

print('H(s)=', Ha)
```

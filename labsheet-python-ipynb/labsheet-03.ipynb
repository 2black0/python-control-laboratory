# Task 1: 

## Modeling for Electronics System - RLC Circuit

```python
import control as ctrl
import numpy as np
import matplotlib.pyplot as plt

# RLC circuit parameters
R = 1.0  # Resistance in ohms
L = 1.0  # Inductance in henrys
C = 1.0  # Capacitance in farads

# Create the transfer function
numerator = [1.0]  # Coefficients of the numerator
denominator = [L*C, R*C, 1]  # Coefficients of the denominator
sys = ctrl.TransferFunction(numerator, denominator)
print(sys)

# Generate the step input signal
t = np.linspace(0, 10, 1000)
u = 1.0  # Scalar step input with a final value of 1.0

# Calculate the step response using control.forced_response()
t_response, y_response = ctrl.forced_response(sys, T=t, U=u)

# Plot the step input and step response on the same graph
plt.plot(t, u * np.ones_like(t), label='Step Input', linestyle='--')
plt.plot(t_response, y_response, label='Step Response')
plt.xlabel('Time (s)')
plt.ylabel('Output Current (I)')
plt.title('Step Response of RLC Series Circuit')
plt.legend()
plt.grid(True)
plt.show()
```

# Task 2:

## Modeling for Mechanical System - Spring Mass Damper

```python
# System parameters
m = 1.0  # mass (kg)
c = 1.0  # damping coefficient (Ns/m)
k = 2.0  # spring constant (N/m)

# Create the transfer function of the spring-mass-damper system
numerator = [1.0]
denominator = [m, c, k]
sys = ctrl.TransferFunction(numerator, denominator)
print(sys)

# Generate the step input signal
t = np.linspace(0, 10, 1000)
u = 1.0  # Scalar step input with a final value of 1.0

# Calculate the step response using control.forced_response()
t_response, y_response = ctrl.forced_response(sys, T=t, U=u)

# Plot the step input and step response on the same graph
plt.plot(t, u * np.ones_like(t), label='Step Input', linestyle='--')
plt.plot(t_response, y_response, label='Step Response')
plt.xlabel('Time (s)')
plt.ylabel('Displacement (m)')
plt.title('Step Response of Spring-Mass-Damper System')
plt.legend()
plt.grid(True)
plt.show()
```

# Task 3:

## Time-domain analysis of control systems using Python

```python
# Define the transfer function
numerator = [2]
denominator = [1, 2, 2]
sys = ctrl.TransferFunction(numerator, denominator)
print(sys)

# Generate the step input signal
t = np.linspace(0, 10, 1000)
u = 1.0  # Scalar step input with a final value of 1.0

# Calculate the step response using control.forced_response()
t_response, y_response = ctrl.forced_response(sys, T=t, U=u)

# Plot the step input and step response on the same graph
plt.plot(t, u * np.ones_like(t), label='Step Input', linestyle='--')
plt.plot(t_response, y_response, label='Step Response')
plt.xlabel('Time (s)')
plt.ylabel('Response')
plt.title('Step Response of the System')
plt.legend()
plt.grid(True)
plt.show()
```

# Task 5:

## Transient Response - Second Order System Transfer Function



```python
# Define the system parameters
omega_n = 2

zeta_underdamped = 0.5
zeta_critically_damped = 1
zeta_overdamped = 2

# Create the transfer functions for the systems
numerator = [omega_n**2]

denominator_underdamped = [1, 2*zeta_underdamped*omega_n, omega_n**2]
underdamped_system = ctrl.TransferFunction(numerator, denominator_underdamped)

denominator_critically_damped = [1, 2*zeta_critically_damped*omega_n, omega_n**2]
critically_damped_system = ctrl.TransferFunction(numerator, denominator_critically_damped)

denominator_overdamped = [1, 2*zeta_overdamped*omega_n, omega_n**2]
overdamped_system = ctrl.TransferFunction(numerator, denominator_overdamped)

print("Underdamped TF:", underdamped_system)
print("Critically damped TF:", critically_damped_system)
print("Overdamped TF:", overdamped_system)
```

```python
# Time vector for the step response
time = np.linspace(0, 10, 1000)

# Compute the step responses of the systems
time_underdamped, response_underdamped = ctrl.step_response(underdamped_system, T=time)
time_critically_damped, response_critically_damped = ctrl.step_response(critically_damped_system, T=time)
time_overdamped, response_overdamped = ctrl.step_response(overdamped_system, T=time)

# Plot the step responses on one canvas
plt.plot(time_underdamped, response_underdamped, label='Underdamped')
plt.plot(time_critically_damped, response_critically_damped, label='Critically Damped')
plt.plot(time_overdamped, response_overdamped, label='Overdamped')

plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title('Step Responses of Different Control Systems')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate system characteristics using control.step_info()
info = ctrl.step_info(underdamped_system)

# Print the system characteristics
print(f"Rise Time: {info['RiseTime']} seconds")
print(f"Settling Time: {info['SettlingTime']} seconds")
print(f"SettlingMin: {info['SettlingMin']} ")
print(f"SettlingMax: {info['SettlingMax']} ")
print(f"Overshoot: {info['Overshoot']}%")
print(f"Undershoot: {info['Undershoot']}%")
print(f"Peak: {info['Peak']}")
print(f"Peak Time: {info['PeakTime']} seconds")
print(f"Steady State Value: {info['SteadyStateValue']}")
```

```python
# Define the transfer function coefficients
numerator = [2]
denominator = [2,4]

# Define the transfer function
sys = control.TransferFunction(numerator, denominator)

# Generate the step input signal
t = np.linspace(0, 10, 1000)
u_step = np.ones_like(t)  # Step input with a final value of 1.0

# Calculate the step response using control.forced_response()
t_response_step, y_response_step = control.forced_response(sys, T=t, U=u_step)

# Steady-State Value for Step Input
steady_state_value_step = y_response_step[-1]
print(f"Steady-State Value for Step Input: {steady_state_value_step:.2f}")

# Desired output for step input with a final value of 1.0
desired_output_step = 1.0

# Steady-State Error for Step Input
steady_state_error_step = desired_output_step - steady_state_value_step
print(f"Steady-State Error for Step Input: {steady_state_error_step:.2f}")

# Plot the step input and step response of the control system on the same graph
plt.plot(t, u_step, label='Step Input', linestyle='--')
plt.plot(t_response_step, y_response_step, label='Step Response')
plt.axhline(y=steady_state_value_step, color='r', linestyle='--', label='Steady-State Value')
plt.xlabel('Time (s)')
plt.ylabel('Response')
plt.title('Step Response of the Control System')
plt.legend()
plt.grid(True)
plt.show()
```
